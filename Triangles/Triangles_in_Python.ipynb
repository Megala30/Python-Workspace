```python
# To print pyramid

P = 5

for i in range(P):
     
    for j in range(P - i - 1):
        print(" ", end="")
        
    for k in range(2 * i + 1):
        print("*", end="")
        
    print()
```

```python
# To print increasing right triangle

rows = 5

for i in range(1, rows + 1):
    
    for j in range(i):
        print("*", end="")
        
    print()
```

```python
# To print decreasing left triangle

rows = 5

for i in range(rows, 0, -1):
     
    for j in range(i):
        print("*", end="")
        
    print()
```

```python
# To print right downward triangle

rows = 5


for i in range(rows):
    
    for j in range(i):
        print(" ", end="")
        
    for k in range(rows - i):
        print("*", end="")
        
    print()
```

```python
# To print left downward triangle

rows = 5

for i in range(rows, 0, -1):
    
    for j in range(i):
        print(" ", end="")
        
    for k in range(rows - i):
        print("*", end="")
        
    print()
```

```python
# To print a double hill 

def print_double_hill(rows):
    
    for i in range(rows):
        
        left_spaces = ' ' * (rows - i - 1)
        
        stars = '*' * (2 * i + 1)
        
        middle_spaces = ' ' * (2 * (rows - i - 1))
        
        print(left_spaces + stars + middle_spaces + stars)

print_double_hill(5)
```

```python
# To print a reverse pyramid

def print_reverse_pyramid(rows):

    for i in range(rows):
        
        spaces = ' ' * i
        
        stars = '*' * (2 * (rows - i - 1) + 1)
        
        print(spaces + stars)

print_reverse_pyramid(5)
```

```python
# To print a butterfly 

def print_butterfly(n):
   
    for i in range(1, n + 1):
    
        stars = '*' * i
        
        spaces = ' ' * (2 * (n - i))
        
        print(stars + spaces + stars)
        
    for i in range(n - 1, 0, -1): 
       
        stars = '*' * i
        
        spaces = ' ' * (2 * (n - i))
        
        print(stars + spaces + stars)

print_butterfly(5)
```

```python
# To print a diamond 

def print_diamond(rows):
   
    
    for i in range(rows):
        spaces = ' ' * (rows - i - 1)
        stars = '*' * (2 * i + 1)
        print(spaces + stars)
        
    for i in range(rows - 2, -1, -1):
        spaces = ' ' * (rows - i - 1)
        stars = '*' * (2 * i + 1)
        print(spaces + stars)

print_diamond(5)
```

```python
# To print a sandglass 

def print_sandglass(rows):
  
    for i in range(rows):
        spaces = ' ' * i
        stars = '*' * (2 * (rows - i - 1) + 1)
        print(spaces + stars)
        
    for i in range(1, rows):
        spaces = ' ' * (rows - i - 1)
        stars = '*' * (2 * i + 1)
        print(spaces + stars)

print_sandglass(5)
```

```python
# To print pascal's triangle 

def print_star_pascals_triangle(rows):
   
    triangle = []
    
    for i in range(rows):
       
        row = [1] * (i + 1)
        
        for j in range(1, i):
            row[j] = triangle[i-1][j-1] + triangle[i-1][j]
            
        triangle.append(row)
        
        left_spaces = ' ' * (rows - i - 1)
        
        row_str = ' '.join('*' if num % 2 != 0 else ' ' for num in row)
        
        print(left_spaces + row_str)

print_star_pascals_triangle(8)
```
